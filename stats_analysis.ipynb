# Dataset Statistics Analysis for Dancetrack

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import json
from statistics import *
import pandas as pd

#set SNS style params
sns.set()

f = open('stats.json')
data = json.load(f)
f.close()
for key in data.keys():
    print(key)
```

```python
# Load Variables
num_sequences = np.array(data['num_sequences'], dtype=np.float32)
num_frames_in_sequence = np.array(data['num_frames_in_sequence'], dtype=np.float32)
num_track_ids_in_sequence = np.array(data['num_track_ids_in_sequence'], dtype=np.float32)
num_pairs_to_annotate_in_sequence = np.array(data['num_pairs_to_annotate_in_sequence'], dtype=np.float32)
num_overlapping_pairs_in_sequence = np.array(data['num_overlapping_pairs_in_sequence'], dtype=np.float32)
```

```python
# Ignore Null Values
num_track_ids_in_sequence_stat = num_track_ids_in_sequence[~np.isnan(num_track_ids_in_sequence)]
num_pairs_to_annotate_in_sequence_stat = num_pairs_to_annotate_in_sequence[~np.isnan(num_pairs_to_annotate_in_sequence)]
num_overlapping_pairs_in_sequence_stat = num_overlapping_pairs_in_sequence[~np.isnan(num_overlapping_pairs_in_sequence)]

# Total stats:
print('Total Number of Frames: ', int(total_frames))
print('Total Number of Pairs to Annotate: ', int(total_pairs_to_annotate))
print('Total Number of Overlapping Pairs: ', int(total_overlapping_pairs))

# Mean stats per sequence:
print("Mean number of frames per sequence:", np.mean(num_frames_in_sequence))
print("STDEV of number of frames per sequence:", np.std(num_frames_in_sequence))
print("Mean number of track IDs per sequence:", np.mean(num_track_ids_in_sequence_stat))
print("STDEV of number of track IDs per sequence:", np.std(num_track_ids_in_sequence_stat))
print("Mean number of pairs to annotate per sequence:", np.mean(num_pairs_to_annotate_in_sequence_stat))
print("STDEV of number of pairs to annotate per sequence:", np.std(num_pairs_to_annotate_in_sequence_stat))
print("Mean number of overlapping pairs per sequence:", np.mean(num_overlapping_pairs_in_sequence_stat))
print("STDEV of number of overlapping pairs per sequence:", np.std(num_overlapping_pairs_in_sequence_stat))
```

#### Plot of Number of Frames in Each Sequence

```python
fig , ax = plt.subplots(figsize=(20, 10))
gfg = sns.barplot(x = np.arange(1,101), y = num_frames_in_sequence, ax=ax)
gfg.set_title( "Number of Frames in Each Sequence")
gfg.set_xlabel( "Sequence Number")
gfg.set_ylabel( "Number of Track IDs")
ax.set_xticks(ax.get_xticks()[1:])
_ = ax.set_xticklabels(ax.get_xticks(), size=8)
plt.show()
```
